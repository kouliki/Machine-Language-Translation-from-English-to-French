```python
def  load_decoder_inputs(decoder_np_vecs='french_train_vecs.npy'):
    # load decoder inputs
    # Indent the lines within the function
    vectorized_title=np.load(decoder_np_vecs)

    decoder_input_data=vectorized_title[:,:-1]
    decoder_target_data=vectorized_title[:,1:]

    print(f'Decoder input shape: {decoder_input_data.shape}')
    print(f'Decoder target shape: {decoder_target_data.shape}')
    return decoder_input_data,decoder_target_data


# load encoder inputs

def load_encoder_inputs(encoder_np_vecs='english_train_vecs.npy'):
  vectorized_body=np.load(encoder_np_vecs)
  encoder_input_data=vectorized_body
  doc_length=encoder_input_data.shape[1]
  print(f'encoder input shape: {encoder_input_data.shape}')
  return encoder_input_data,doc_length
```

```python
def load_text_processor(fname='english_pp.pkl'):
  with open(fname,'rb') as f:
    pp=dpickle.load(f)

  num_tokens=max(pp.id2token.keys())+1
  print(f'Size of vocabulary for {fname}: {num_tokens:,}')
  return num_tokens,pp
```

```python
from keras.models import Model
from keras.layers import Input,LSTM,Dense,GRU,Embedding,Bidirectional,BatchNormalization
from keras import optimizers
```

```python
from os import name
latest_dim=300
latent_dim = latest_dim # Initialize latent_dim

encoder_inputs=Input(shape=(doc_length,),name='encoder_input')

x=Embedding(num_encoder_tokens,latent_dim,name='Body-Word-Embedding',mask_zero=False)(encoder_inputs)
x=BatchNormalization(name='Encoder-BatchNormalization')(x)

# Assign the output of GRU to a variable
state_h = GRU(latent_dim,return_state=True,name='Encoder-GRU')(x)

encoder_model=Model(inputs=encoder_inputs,outputs=state_h,name='Enoder-Model')
seq2seq_encoder_out=encoder_model.predict(encoder_input_data)
```

```python
# Decoder Model
decoder_input=Input(shape=(None,),name='decoder_input')

dec_emb=Embedding(num_decoder_tokens,latent_dim,name='Target-Word-Embedding',mask_zero=False)(decoder_input)
dec_bn=BatchNormalization(name='Decoder-BatchNormalization')(dec_emb)

# Use the encoder output as the initial state for the decoder
# Note: We don't call 'predict' here, we directly pass the encoder model's output tensor
decoder_gru=GRU(latent_dim,return_sequences=True,return_state=True,name='Decoder-GRU')
# Only use the hidden state as the initial state
decoder_gru_output,_=decoder_gru(dec_bn,initial_state=encoder_model(encoder_inputs)[0])

x=BatchNormalization(name='Decoder-BatchNormalization')(decoder_gru_output)

decoder_dense=Dense(num_decoder_tokens,activation='softmax',name='Decoder-Dense')
decoder_output=decoder_dense(x)
```

```python
# Seq2seq_Model

seq2seq=Model([encoder_inputs,decoder_input],decoder_output)
seq2seq.compile(optimizer=optimizers.Nadam(lr=0.001),loss='sparse_categorical_crossentropy',metrics=['accuracy']) # Use 'seq2seq' instead of 'seq2seq_Model'
```

```python
def extract_encoder_model(model):
  encoder_model=model.get_layer('Encoder-Model')
  return encoder_model
```

```python
def extract_decoder_model(model):

    latent_dim=model.get_layer('Decoder-BatchNormalization').output_shape[-1]
    decoder_inputs=model.get_layer('decoder_input').input
    dec_emb=model.get_layer('Decoder-Word-Embedding')(decoder_inputs)
    dec_bn=model.get_layer('Decoder-BatchNormalization')(dec_emb)

    gru_inference_state_input=Input(shape=(latent_dim,), name='hidden_State_input')
    gru_out,gru_state_out=model.get_layer('Decoder-GRU'([dec_bn,gru_inference_state_input]))

    #Reconstruct dense layers

    dec_bn2=model.get-layer('Decoder-Batchnorm-2')(gru_out)
    dense_out=model.get_layer('Decoder-Dense')(dec_bn2)

    decoder_model=Model([decoder_inputs,gru_inference_state_input],dense_out)
    return decoder_model





```

```python
!pip install nltk
import nltk
import nltk.translate.bleu_score as bleu
import nltk.translate.meteor_score as meteor

# Download the WordNet resource
nltk.download('wordnet')

def calculate_accuracy(predictions, references):
    """
    Calculates accuracy metrics for seq2seq model predictions.

    Args:
        predictions: A list of predicted output sequences.
        references: A list of corresponding true target sequences.

    Returns:
        A dictionary containing BLEU and METEOR scores.
    """

    bleu_score = bleu.corpus_bleu(references, predictions)
    meteor_score = meteor.meteor_score(references, predictions[0])  # Assuming predictions is a list of lists

    return {
        "BLEU": bleu_score,
        "METEOR": meteor_score
    }

# Example usage:
predicted_sequences = [["this", "is", "a", "prediction"], ["another", "predicted", "sequence"]]
true_sequences = [["this", "is", "the", "reference"], ["another", "true", "sequence"]]

accuracy_metrics = calculate_accuracy(predicted_sequences, true_sequences)
print(accuracy_metrics)
```
